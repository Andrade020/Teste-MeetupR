```r
y=NULL
x=NULL
Cn<-NULL
Ch<-NULL 
Ctot<-NULL
Cn[1]=5000
Ch[1]=5000
Ccons=10000
Ctot[1]=Ch[1]+Cn[1]
x[1]=1
for (i in 2:30){
  x[i]= i
  Ch[i]=Ch[i-1]+ 3*i
  if(i>10 && i<13) Ch[i]=Ch[i-1]-50*i
  if(i>12 && i<15) Ch[i]=4200-100*i
  if(i==21) Ch[i]=2900
  if(i==22) Ch[i]=4200
  Cn[i]= Cn[i-1]+ ( (Ccons-Ctot[i-1])/Ccons)*(1- Cn[i-1]/Ccons)*Cn[i-1]
  Ctot[i]= (Ch[i]+Cn[i])+ rnorm(1)*100
  
  
}
plot(x,Ctot,type="l" , xlab = "time", ylab = "amount of substance")
```
